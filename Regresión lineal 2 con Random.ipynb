```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

Se construira una serie de datos que cumplan los crterios para poder utilizar la regresi칩n lineal en este ejercicio. recordemos que la ecuacon de este met칩do corresponde a una funcion lineal con pendiente m

```python
#Se fija un estado aleatorio para poder tener reproducibilidad
#dicho de tora manera se "fija una semilla", al usar
#random.seed(Value)siempre se nos generar치 la misma serie de
#numeros aleatorios
np.random.seed(48)
#X-atributos, y-variable a predecir
X = 2 * np.random.rand(100,1) #genera un array de forma 
                        #especificada con valores aleatorios,
                        #100 filas 1 columna en el intervalo
                        # [0,2)
# numpy.random.randn genera muestras de la distribuci칩n normal, 
#mientras que numpy.random.Rand de unifrom (en el rango [0,1])
y = 4 + 3 * X + np.random.randn(100,1)
#el ultimo termino que se usa es para que no quedan sobre
#la recta todos nuestros valores, lo que los hace "variables"
#intersswecion con y=4 pendiente=3
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Conjunto de datos lineales generados aleatoriamente')
plt.grid() #me cuadricula
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
lr = linear_model.LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test, Y_pred, color = 'red', linewidth = 2)
plt.title('Regresion lineal simple')
plt.show()
```
