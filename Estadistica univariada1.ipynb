## 4. Tablas de frecuencia
Para la solucion del caso de uso es necesario identificar una posible hipótesis para validad a traves de las herramientas que se revisaran en las siguientes lineas.

La tabla de frecuencia es una heramienta que nos permite ordenar los datos, de manera que se presentan numericamente las caracteristicas de la distribucion de un conjunto de datos o muestra.

### Hipotesis:
La categoria de celulares es la que mas producto vende, se quiere validar ya que quiere lanzar una campaña de descuento para traer mas clientes a la empresa, sin embargo se desea conocer cual es la categoria en la que se invertirá mas publicidad.

```python
##Tabla de frecuencia univariada
#esto para ver que categoria se vende mas
#Frecuencia absoluta

frec_data = data['nombre_categoria_producto'].value_counts().reset_index()
frec_data.head(4)
```

```python
#frecuencia relativa
frec_data['porcentaje_nombre_categoria_producto'] = frec_data['nombre_categoria_producto'].apply(lambda x: x/frec_data['nombre_categoria_producto'].sum())
frec_data.head(5)
```

```python
#la frecumencia realiva y absoluta acumulada usando cumsum()
frec_data['frec absoluta acumulada'] = frec_data['nombre_categoria_producto'].cumsum()
frec_data['frec relativa acumulada'] = frec_data['porcentaje_nombre_categoria_producto'].cumsum()
```

```python
frec_data
```

```python
## frecuencia bivariada
#tambla que nos muestra la frecuencia por ciudad de venta y categoria del producto
pd.crosstab(index = data['nombre_categoria_producto'], columns = data['ciudad_vendedor'])
```

### Análisis:
De la tabla de frecuencias podemos observar que en frecuencia aucmulada los productos de tecnologia son los más vendidos

## 5. Tendencia central (moda, media y mediana)

### Hipótesis: 
los precios en el comercio son mayores a 500, ya que el estudio de marketing arrojó que los clientes más recurrentes compran mayormente productos más económicos (debajo de 500) por lo que es necesario analizar si en promedio la compañia tiene precios mas competitivos.

```python
#media o poromedio de precios
data['precio'].mean()
```

```python
#Mediana. Valor central de la distribucion de los datos
data['precio'].median()
```

```python
#Moda. Los precios que mas se repiten 
data['precio'].mode()
```

### Análisis: 
Es posible analizar el promedio y se observa que se encnuentra por debajos de los 500, localizandoce en los 120, por lo que en promedio se compran productos cuyo precio se encuentra por debajo de los 120 pesos, dentro de los cuales los productos con precio menor a 60 son lo smas comprados.

## 6. Medidas de localización
### Hipótesis:
la mitad (50%) de los productos pesa mas de 500g, ya que la empresa de mensajeria quiere realizar un descuento si los productos pesan menos de 500g, asi la compañia podria bajar los precios de envio de la mitad de los productos que se venden.

```python
##Cuantiles
# calcular el 20%, 50% y 70% del peso de los productos 
data['peso_g_producto'].quantile([0.20, 0.50, 0.70])
```

### Análisis:
podemos ver que el cuantil 50% está por arriba de los 500g, por lo que mas de la mitad de los productos pesan mas de 500g, lo cual nos rechaza la hipótesis.

## 7. Medidas de disperción
### Hipótesis: 
La variacion de las longitudes de los datos es mayor a 20cm, ya que las de empresas de mensajeria que tienen cubiculos de transporte de mensajeria individual de producto de longitud de 20 cm son mas economicas, asi se podria cambiar a estas empresas y ahorrar en costos de envio.

```python
##Desviacion estandar
#localizar la desviacion estandar de la ongitud de los productos de la mucestra
#representa que tan lejos de lamedias e encuentran las observaciones
data['longitud_cm_producto'].std()
```

```python
##Rango intercuartilico
#encontrar el rango intercuartilico de la longitud de los productos
#es la diferencia entre el tercer y primer cuartil
data['longitud_cm_producto'].quantile(0.75) - data['longitud_cm_producto'].quantile(0.25)
```

```python
##Rango 
#muestra la diferencia del producto mas grande con el mas chico
data['longitud_cm_producto'].max() - data['longitud_cm_producto'].min()
```

```python
## Coeficiente de variacion 
# diviosn de la desciacion estander entre la sobre la media 
data['longitud_cm_producto'].std() / data['longitud_cm_producto'].mean()
```

### Analisis: 
A pesar que la desviacion estandar y el rango intercuartilico son menos a 25, se puede analizar ocn el rando que la variacion entre el máximo y el mínimo es de 91cm, por lo tanto se podria migrar al rededor de 50% de los preductos de las empresas de mensajeria con el cubivulo de 20cm y darle al cliente un descuento en el envio.
